```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame(index=range(1, 51, 1))
df["I"]=df.index.map(lambda x: -1/(2*(np.log(x))**2)+1/(2*(np.log(2))**2))
```

```python
fig, ax=plt.subplots(figsize=[7, 5])
df.plot(ax=ax)

ax.axvline(2, linestyle="--", linewidth=0.7, color="gray", label="x=2")
ax.axhline(1/(2*(np.log(2))**2), linestyle="--", linewidth=0.7, color="red", label=f"x={np.round(1/(2*(np.log(2))**2), 3)}")

ax.axhline(0, linewidth=0.7, color="black")
#ax.axvline(0, linewidth=0.7, color="black")

ax.legend()
```

```python
df2=pd.DataFrame(index=range(1, 51, 1))
df2["1/log^2x"]=df.index.map(lambda x: 1/np.log(x)**2)

fig, ax=plt.subplots(figsize=[7, 5])
df2.plot(ax=ax)
```
