```python
# importing local ant module
import sys
import os
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path+"/")
import ant

# normal imports
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(index=[x*0.1 for x in range(1, 100, 1)])
df.index.rename('t', inplace=True)
df['|(1-3logt)/t^4|']=df.index.map(lambda t: np.abs(t**(-4)*(1-3*np.log(t))))
df.plot()
```

```python
df2=pd.DataFrame(index=[x*0.1 for x in range(1, 100, 1)])
df2.index.rename('t', inplace=True)
df2['y']=df2.index.map(lambda t: t**(-4))
df2.plot()
```

```python
df3=pd.DataFrame(index=[x*0.1 for x in range(1, 1000, 1)])
df3.index.rename('t', inplace=True)
# df3['logt']=df3.index.map(lambda t: np.log(t))
df3['1/t']=df3.index.map(lambda t: 1/t)
df3['logt/t']=df3.index.map(lambda t: np.log(t)/t)
df3['1']=1
df3.plot()
```

```python
df4=pd.DataFrame(index=[x*0.1 for x in range(50, 1000, 1)])
df4.index.rename('t', inplace=True)
# df3['logt']=df3.index.map(lambda t: np.log(t))
df4['1/t^2']=df4.index.map(lambda t: 1/(t**2))
df4['1/t']=df4.index.map(lambda t: 1/t)
#df4['1']=1
df4.plot()
```

```python
df5=pd.DataFrame(index=[x*0.1 for x in range(50, 1000, 1)])
df5.index.rename('t', inplace=True)
df5['t^1/2']=df5.index.map(lambda t: t**(1/2))
df5['t^-1/2']=df5.index.map(lambda t: t**(-1/2))
df5['t^1/2logt']=df5.index.map(lambda t: np.log(t)*t**(1/2))
df5.plot()
```
