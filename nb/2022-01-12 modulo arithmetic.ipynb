```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame(index=range(0, 10, 1))
df1["3^n"]=df1.index.map(lambda n: 3**n)
df1["mod5"]=df1["3^n"]%5
df1
```

```python
df2=pd.DataFrame(index=range(0, 10, 1))
df2["n^3"]=df2.index.map(lambda n: n**3)
df2["n^3 mod5"]=df2["n^3"]%5
df2["cumsum"]=df2["n^3"].cumsum()
df2["cumsum mod5"]=df2["cumsum"]%5
df2
```

```python
df3=pd.DataFrame(index=range(0, 10))
df3.index.rename("n", inplace=True)
df3["n!"]=df3.index.map(lambda n: np.math.factorial(n))
df3["mod5"]=df3["n!"]%5
df3.T # T for transpose
```

```python
df4=pd.DataFrame(
    index=range(1, 14, 1),
    data={"x":[0]+[2**n for n in range(0, 12)]}
)
df4["mod13"]=df4.x%13
df4.T
```

```python
df5=pd.DataFrame(
    index=range(0, 13),
    data={"n^2":[n**2 for n in range(0, 13)]}
)
df5.index.rename("n", inplace=True)
df5["mod13"]=df5["n^2"]%13
df5.T
```

```python
df6=pd.DataFrame(
    index=range(0, 13),
    data={"n^3":[n**3 for n in range(0, 13)]}
)
df6.index.rename("n", inplace=True)
df6["mod13"]=df6["n^3"]%13
df6.T
```

```python
df7=pd.DataFrame(
    index=range(1000, 1025, 1),
)
df7.index.rename("n", inplace=True)
df7["mod13"]=df7.index%13
display(df7.T)
print("Unique residues:", df7["mod13"].unique())
```
